# Dataset Information

Source: Kaggle [link](https://www.kaggle.com/datasets/belbino/indian-banking-transactions-20192024)

Year / Timeline: Data collected during 2019 to 2024

Size: 550000

Dataset Description:

1.  transaction_id – Unique identifier for each transaction
2.  customer_id – Unique ID assigned to each customer
3.  transaction_date – Date on which the transaction occurred
4.  transaction_time – Time of the transaction
5.  account_type – Type of bank account (Savings, Current, etc.)
6.  transaction_type – Nature of transaction (Deposit, Withdrawal, Transfer, Payment)
7.  transaction_amount – Amount involved in the transaction
8.  transaction_direction – Indicates whether the transaction is credit or debit
9.  account_balance – Available balance after the transaction
10. merchant_category – Category of merchant (Retail, Grocery, Online, etc.)
11. state – Customer’s location/state
12.  credit_score – Creditworthiness score of the customer
13.  has_loan – Indicates whether the customer has a loan (Yes/No)
14.  loan_type – Type of loan (Home, Personal, Auto, etc.)
15.  emi_amount – Monthly installment amount (if applicable)
16.  transaction_status – Status of transaction (Success, Failed, Pending)
17.  channel – Mode of transaction (ATM, Online, Mobile Banking, Branch)
18.  kyc_status – KYC verification status (Verified / Not Verified)
19. is_fraud – Fraud indicator (1 = Fraud, 0 = Not Fraud)
20. transaction_hour – Hour of the day when transaction occurred

# Type of Analysis

**Descriptive Analysiss**

This involves summarizing the dataset to understand basic features such as total transactions, average transaction amount, distribution of transaction types, and fraud rate. It provides a clear overview of the data using statistics and visualizations.

**Diagnostic Analysis**

This focuses on identifying the reasons behind observed patterns. For example, analyzing why fraud occurs more in certain transaction types, channels, or time periods, and understanding the relationship between credit score, account balance, and fraud.

**Predictive Analysis**
This involves using historical data to predict future outcomes. For instance, basic statistical methods can be used to estimate the likelihood of a transaction being fraudulent based on factors like transaction amount, time, and customer profile.

**Prescriptive Analysis**

This provides recommendations based on insights. For example, suggesting stricter monitoring for high-risk transactions, improving KYC processes, or enhancing fraud detection systems to reduce financial risk.

## Stage 2 – Data Cleaning and Pre-processing



```python
# Handle Missing Values
# Fill 'loan_type' with 'No loan' where 'has_loan' is 0 and 'loan_type' is NaN
# Fill 'loan_type' with 'No loan' where 'has_loan' is 1 and 'loan_type' is NaN
df.loc[(df['has_loan'] == 0) & (df['loan_type'].isna()), 'loan_type'] = 'No loan'
df.loc[(df['has_loan'] == 1) & (df['loan_type'].isna()), 'loan_type'] = 'Unkwon'
```

```python
print(df.isnull().sum())
```

```python
df.isnull().sum()
```

```python
# Handle Duplicates
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
df.nunique()
```

```python
df.describe()
```

```python
# Check Skewness
print("Skewness:", df['transaction_amount'].skew())
```

```python
df['log_transaction_amount'] = np.log1p(df['transaction_amount'])
```

```python
df.dtypes
```

```python
# Convert Data Types
df['transaction_date'] = pd.to_datetime(df['transaction_date'], errors='coerce', dayfirst=True)
```

```python
df['has_loan'] = df['has_loan'].astype(bool)
df['is_fraud'] = df['is_fraud'].astype(bool)
```

```python
df.dtypes
```

### Feature Engineering: Derived Fields

```python
# Create 'time_of_day_category'
def get_time_of_day(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 16:
        return 'Afternoon'
    elif 16 <= hour < 20:
        return 'Evening'
    else:
        return 'Night'

df['time_of_day_category'] = df['transaction_hour'].apply(get_time_of_day)
df.head()
```
